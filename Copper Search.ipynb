```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.edge.options import Options

# Set the path to your Microsoft Edge WebDriver executable
edge_driver_path = r'C:\Program Files\msedgedriver.exe'  # Replace with the actual path to msedgedriver.exe

# Initialize the EdgeOptions object and specify the driver path
edge_options = Options()
edge_options.binary_location = edge_driver_path

# Initialize the Edge WebDriver with the specified options
driver = webdriver.Edge()

# Open a new tab
driver.execute_script("window.open('');")
driver.switch_to.window(driver.window_handles[1])

# Navigate to a website (in this case, the default search engine)
driver.get("https://www.bing.com")

# Find the search bar element by its name attribute using the By.NAME locator
search_bar = driver.find_element(By.NAME, "q")

# Type "copper" into the search bar
search_bar.send_keys("copper")

# Press Enter to perform the search
search_bar.send_keys(Keys.RETURN)

# You can add a delay here to see the results if needed
# import time


```
